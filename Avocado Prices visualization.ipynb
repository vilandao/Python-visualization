# Avocado Prices visualization with Seaborn
------------------------

### Content: 

This dataset was downloaded from Kaggle. The data was obtained from the Hass Avocado Board website and was described as:

The database represents weekly 2015-2018 retail scan data for National retail volume (units) and price. Retail scan data comes directly from retailers’ cash registers based on actual retail sales of Hass avocados. Starting in 2013, the database reflects an expanded, multi-outlet retail data set. Multi-outlet reporting includes an aggregation of the following channels: grocery, mass, club, drug, dollar and military. The Average Price (of avocados) in the table reflects a per unit (per avocado) cost, even when multiple units (avocados) are sold in bags. The Product Lookup codes (PLU’s) in the table are only for Hass avocados. Other varieties of avocados (e.g. greenskins) are not included in this table.

Some relevant columns in the dataset:

* **Date** - The date of the observation
* **AveragePrice** - The average price of a single avocado
* **Total Volume** - Total number of avocados sold
* **4046** - Total number of avocados with PLU 4046 sold
* **4225** - Total number of avocados with PLU 4225 sold
* **4770** - Total number of avocados with PLU 4770 sold
* **Total Bags** - Total bags of avocado for the listed time period
* **type** - conventional or organic
* **year** - the year
* **region** - the city or region of the observation

### Data exploratory and Visualization

```python
# Import working liabraries:
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Load the 'avocado' dataset using pandas:
df = pd.read_csv('avocado.csv', index_col=0)
df.head()
```

```python
# Perform exploratory analysis on df:
df.describe(include='all')
```

```python
# Check for missing values:
df.isnull().sum()
```

```python
df.info()
```

```python
# Split date, time, and year into 3 different columns and drop the duplicated 'year' column:
df['Year'], df['Month'], df['Day']= df['Date'].str.split('-').str
df.drop(['year'], axis=1, inplace=True)
df.head()
```

#### Task 1 : Choose one variable and plot that variable 4 different ways

```python
# Pick 'AveragePrice' for analysis as I'd like to look into the prices of avocado over time
plt.figure(figsize=(9,6))

sns.set(style='darkgrid')
sns.set(font_scale=1.5)
sns.scatterplot(x='Year', y='AveragePrice', data=df, hue='type')
plt.title('Avocado Average Price by type from 2015-2018')
plt.ylabel('Average Price')
plt.show()
```

```python
plt.figure(figsize=(9,6))

sns.set(style='whitegrid')
sns.set(font_scale=1.5)
sns.distplot(df['AveragePrice'], kde_kws={"color": "k", "lw": 2, "label": "KDE"}, color='purple')
plt.title('Avocado Average Price Distribution')
plt.xlabel('Average Price')
plt.show()
```

```python
plt.figure(figsize=(9,6))

sns.set(style='darkgrid')
sns.set(font_scale=1.5)
sns.lineplot(x='Month', y='AveragePrice',hue='type', style='type', markers=True, data=df, markersize=10)
plt.title('Avocado Average Price by type')
plt.xlabel('Month of the year')
plt.ylabel('Average Price')
plt.show()
```

```python
plt.figure(figsize=(12,8))

sns.set(font_scale=1.5)
sns.boxplot(x='Month', y='AveragePrice', hue='type', data=df, palette="Set2")
plt.title('Avocado Average Price distribution by type')
plt.ylabel('Average price')
plt.show()
```

#### Task 2: Choose two continuous variables, and plot them 3 different ways

```python
# Choose the 'AveragePrice' and 'Total Volume' variables for analysis:
plt.figure(figsize=(10,6))

sns.set(font_scale=1)
sns.scatterplot(x='Total Bags', y='AveragePrice', data=df, hue='type')
plt.title('Avocado Total Bags and Average Price')
plt.ylabel('Average Price')
plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.set(font_scale=1)
sns.lmplot(x='Total Volume', y='Total Bags', hue='Year', col='Year', data=df, height=6, aspect=.6, x_jitter=.1)
plt.tight_layout()
plt.title('Avocado Volume and Bags comparison 2015-2018')
```

```python
sns.set(style="ticks", color_codes=True)
sns.pairplot(df.iloc[:,7:11],height=3, hue='type')
plt.title('Avocado volume comparison')
plt.show()
```

#### Task 3: Choose one continuous variable and one categorical variable, and plot them six different ways

```python
# Filter to only include 2018 avocado data as I'm interested in looking at the most current year:
df = df[df['Year']=='2018']

# Filter out regions that don't make sense:
filter =df['region']!='TotalUS'
filter =df['region']!='SouthCentral'
filter =df['region']!='West'
df=df[filter]
df['region'].unique()
```

```python
sns.set(style='darkgrid')
sns.set(font_scale=1.2)
sns.factorplot('AveragePrice','region',data=df,
                   hue='type',
                   size=13,
                   aspect=1,
                   palette='Reds',
                   join=False)
plt.title('Average avocado prices by region in 2018')
plt.show()
```

```python
plt.figure(figsize=(24,10))

sns.set(font_scale=2)
sns.barplot(x='region',y='AveragePrice', data=df, palette="Blues_d")

plt.xticks(rotation=90)
plt.xlabel('Region')
plt.ylabel('Average Price')
plt.title('Average Price of Avocado According to Region')
```

```python
plt.figure(figsize=(24,10))

sns.set(font_scale=2)
sns.stripplot(x='region',y='AveragePrice', data=df, size=7, jitter=True,linewidth=1, palette="Greens_d")

plt.xticks(rotation=90)
plt.xlabel('Region')
plt.ylabel('Average Price')
plt.title('Average Price of Avocado According to Region')
```

```python
plt.figure(figsize=(24,10))

sns.set(font_scale=2)
sns.boxplot(x='region',y='AveragePrice', data=df)

plt.xticks(rotation=90)
plt.xlabel('Region')
plt.ylabel('Average Price')
plt.title('Average Price of Avocado According to Region')
```

```python
plt.figure(figsize=(25,10))

sns.set(font_scale=2)
sns.pointplot(x='region',y='AveragePrice', hue='type', data=df)

plt.xticks(rotation=90)
plt.xlabel('Region')
plt.ylabel('Average Price')
plt.title('Average Price of Avocado According to Region')
```

```python
plt.figure(figsize=(10,25))

sns.set(font_scale=1)
sns.stripplot(x='AveragePrice',y='region', data=df, size=7, jitter=True,linewidth=1)
sns.violinplot(x='AveragePrice', y='region', data=df, inner=None, color=".8")

plt.xticks(rotation=90)
plt.xlabel('Region')
plt.ylabel('Average Price')
plt.title('Average Price of Avocado According to Region')
```
